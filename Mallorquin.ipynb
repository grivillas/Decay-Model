```python
#!/usr/bin/env python2
# -*- coding: utf-8 -*-
"""
Created on Thu May  2 13:05:00 2019

@author: German Rivillas
"""


import numpy as np
import matplotlib.pyplot as plt # IMPORTANDO LIBRERIA PARA GRAFICAR
import math # IMPORTANDO LIBRERÌA DEL MODULO MATH

# ********* DATOS DE ENTRADA *********
dt=3600
tsim=24
Niter=int((tsim*60*60)/dt)
k=0.0001;
#NivC=100;

m=1;

# ********* TIPO DE ESQUEMA NUMERICO *********

esq=int(input("tipo de esquema"))

if esq==1:
         tder=-dt*k
elif esq== 2:
         tder=(1+dt*k)
         
# ********* CONDICIÒN DE ESTABILIDAD NUMERICA1 *********
if tder<=0:
    print("reducir el paso de tiempo")

# DEFINICION DE VECTORES EN LAS VARIABLES 

vcon=[]
pasot=[]
t=[]
cval=[]
ee=[]
#NivC=[]
#NivC=np.zeros((1,24))
#NivC[0,0]=100
NivC=np.arange(100,76,-1)

# CALCULO DE LA TASA DE DECAIMIENTO 

for m in range(0,Niter,1):
    pasot.append(NivC[m]*tder)
    t.append(m*dt)
    ee.append(math.exp(-k*t[m]))
    cval.append(NivC[m]*ee[m])
    NivC[m]=cval[m]
    
# GRAFICO DE LA TASA DE DECAIMIENTO
    
plt.plot(t,NivC)
plt.xlabel("Tiempo [h]")
plt.ylabel("Decaimiento [%]")
plt.title("Tasa Decaimiento Contaminantes Cienaga Mallorquin")
plt.grid(True)
plt.show()
```
